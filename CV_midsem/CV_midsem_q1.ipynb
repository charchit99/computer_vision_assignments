```python
import numpy as np
from matplotlib import pyplot as plt

def convx(I):
    Ix = np.zeros((I.shape[0]-2, I.shape[1]-2))
    for i in range(I.shape[0]-2):
        for j in range(I.shape[1]-2):
            Ix[i][j] = -I[i+1][j] + I[i+1][j+2]
    return Ix

def convy(I):
    I = I.T
    Ix = np.zeros((I.shape[0]-2, I.shape[1]-2))
    for i in range(I.shape[0]-2):
        for j in range(I.shape[1]-2):
            Ix[i][j] = -I[i+1][j] + I[i+1][j+2]
    return Ix.T

def getH(Ix, Iy):
    H = np.zeros((2, 2))
    sm = 0
    for i in range(Ix.shape[0]):
        for j in range(Ix.shape[1]):
            sm = sm+Ix[i][j]*Ix[i][j]
    H[0][0]=sm
    sm = 0
    for i in range(Ix.shape[0]):
        for j in range(Ix.shape[1]):
            sm = sm+Iy[i][j]*Iy[i][j]
    H[1][1]=sm
    sm = 0
    for i in range(Ix.shape[0]):
        for j in range(Ix.shape[1]):
            sm = sm+Ix[i][j]*Iy[i][j]
    H[0][1]=sm
    H[1][0]=sm
    return H

def printMatrix(I):
    for i in range(I.shape[0]):
        for j in range(I.shape[1]):
            print(I[i][j], end = " ")
        print("\n")
    print("----------------------")

def getscore(I):
    I = np.array(I)
    Ix = convx(I)
    Iy = convy(I)
    H = getH(Ix, Iy)
    print("I")
    printMatrix(I)
    print("Ix")
    printMatrix(Ix)
    print("Iy")
    printMatrix(Iy)
    print("H")
    printMatrix(H)
    detH = np.linalg.det(H)
    traceH = np.trace(H)
    w, v = np.linalg.eig(H)
    score = detH - 0.05*traceH*traceH
    print("Eigen values = ", w)
    print("Harris response = ", score)
    if traceH!=0:
        print("Class wala formula = ", detH/traceH)
    print()
    print("- - - - - - -")
    print()
```

### Question 2: Given a light source, a solid planar triangle, and a planar ground surface find the shadow of the triangle on the planar ground surface using homography.

#### Answer 2: 

To find the homography matrix we require $R_{world}^{camera}$, $t_{camera}^{world}$, and K (camera matrix). 

We already know $t_{camera}^{world}$ as $X_{camera}^{w}$ (camera's world coordinate).

We assume the following for the camera matrix:  

$K = \left[\begin{array}{ccc}
f & 0 & 0 \\
0 & f & 0 \\
0 & 0 & 1 \\
\end{array}\right]$


For calculating $R_{world}^{camera}$, we first find the vector normal to the image plane, $\vec{N}$. We need to rotate this vector about some axis $\vec{Q}$ by angle $\theta$ such that $\vec{N}$ becomes parallel to the z axis of the world.

$\vec{N} = \vec{AB} \times \vec{BC}$, where $\vec{AB}$ and $\vec{BC}$ are the sides of the triangle.  

f = D = distance of the origin from the plane = $\hat{N} \cdot (\vec{A} - \vec{C})$, where C is the camera vector wrt world origin.  

$\theta = cos^{-1}\dfrac{\vec{N}.\vec{Z}}{|\vec{N}||\vec{Z}|}$

$\vec{Q} = \vec{N} \times \vec{Z}$, where $\vec{Z} = \hat{k}$ (i.e., the z axis).  

$\displaystyle{ \mathbf{R} = \mathbf{I} + (\sin\theta) \mathbf{K} + (1-\cos\theta)\mathbf{K}^2 }$, where K is the skew symmetric matrix of $\vec{Q}$.

Alternatively, for finding $R_{world}^{camera}$ we can use the `LookAt method`, which says:

$R = \left[\begin{array}{ccc}
Q_x & Q_y & Q_z \\
0 & 0 & 1 \\
\hat{N}_x & \hat{N}_y & \hat{N}_z \\
\end{array}\right]$

$t = P_4 = \left[\begin{array}{ccc}
Q \cdot P\\
Z \cdot P\\
N \cdot P\\
\end{array}\right]$

$H = K * [R' | t]$, here R' = R[:, :2] and H is the world to image homography.

Therefore, we can clearly find the location of the shadow points given the Homography matrix and image points. An extension to this problem could be to find the shadow points given the points of the traingle in the world frame.

We have the following equation:  
$x_i = H * X_o$  

from the projection matrix we can write:  
$x_i = P * X_i$, where $X_i$ contains the 3D triangle points in the world frame.  

$\implies X_o = H^{-1} * P * X_i$


On solving the above equation we get the following:

$\implies X_o = W * X_i$, where W only depends on the world coordinates of the light source! 

$W = \left[\begin{array}{ccc}
z_{light} & 0 & -x_{light} & 0\\
0 & z_{light} & -y_{light} & 0\\
0 & 0 & 0 & 0 \\
0 & 0 & -1 & z_{light} \\
\end{array}\right]$

Now we can directly use this formula for calculating the shadow points, the following code does that.

```python
def getH(xt, yt, zt):
    H = [[zt, 0, -xt, 0], 
         [0, zt, -yt, 0],
         [0,  0,   0, 0],
         [0,  0,  -1,zt]]
    return np.array(H)

def transform(H, X):
    Y = H @ X.T
    Y = Y/Y[-1]
    return Y[:3, :].T

side, ht = 4, 2
pos = np.array([[side/2, side/2, ht, 1], [-side/2, side/2, ht, 1], [side/2, -side/2, ht, 1]])

lightSource = np.array([0, 10, 2 * ht])
Hmatrix = getH(lightSource[0], lightSource[1], lightSource[2])
shadowPts = transform(Hmatrix, pos)
```

```python
ax = plt.axes(projection='3d')
print(shadowPts)
ax.scatter3D(shadowPts[:, 0], shadowPts[:, 1], shadowPts[:, 2], c = "black")#, c=zdata)
ax.scatter3D(pos[:, 0], pos[:, 1], pos[:, 2], c = "red")
ax.scatter3D(lightSource[0], lightSource[1], lightSource[2], c = "orange")
```

#### Why I found this problem interesting  
The problem itself is very interesting. Here in the solution for the problem the object itself becomes the image and the shadow (which should ideally have been the image) is the object. The homography gives the direct mapping from the obeject to the image and its inverse and can work even when the image or the light source position has been changed which is what makes this question unique.

```python
def generate_A(x_homogeneous, X_homogeneous):
    n = x_homogeneous.shape[0]
    x, y = x_homogeneous[:, 0], x_homogeneous[:, 1]
    X, Y, Z = X_homogeneous[:, 0], X_homogeneous[:, 1], X_homogeneous[:, 2]

    A1 = np.c_[-X, -Y, -Z, -np.ones(n), np.zeros(n), np.zeros(n), np.zeros(n), np.zeros(n), np.multiply(x, X), np.multiply(x, Y), np.multiply(x, Z), x]
    A2 = np.c_[np.zeros(n), np.zeros(n), np.zeros(n), np.zeros(n), -X, -Y, -Z, -np.ones(n), np.multiply(y, X), np.multiply(y, Y), np.multiply(y, Z), y]

    # Append the rows of A1 and A2 to A in a sequential manner

    A = np.zeros((2 * n, 12))
    A[::2, :], A[1::2, :] = A1, A2
    return A

def DLT(x_homogeneous, X_homogeneous):
    A = generate_A(x_homogeneous, X_homogeneous)
    U, D, Vt = np.linalg.svd(A)
    p_vector = Vt[-1, :]            # get the last row of Vt, equivalent to getting last col of V
    P = p_vector.reshape((3, 4))    # transforming the p_vector to a matrix P
    return P

P = DLT(x, X)
print(P)

P = P/P[-1, -1]

print("The homogeneous form looks like:\n", P)

def ret_RKT(P):

    # P = K * [R | -R * X0] = [KR | -KRX0] = [H|h]
    H, h = P[:, :3], P[:, -1]

    # Computing pseudoinverse (for cases when inverse doesn't exist)
    H_inverse = np.linalg.pinv(H)
    X0 = - H_inverse @ h

    # QR decomposition decomposes a matrix into a triangular and rotation matrix
    # QR(M) = (Rot, Tri), where M = Rot * Tri
    # H = K * R     => H^-1 = R^T * K^-1
    # If K is triangular then so is K^-1
    # And given the order of decomposition we calculate the QR decompostion of H^-1 instead of that for H.

    R_transpose, K_inverse = np.linalg.qr(H_inverse)
    K = np.linalg.pinv(K_inverse)

    # The H matrix is homogenous, => so is the calibration matrix
    # Due to its homogeneity, we normalize it

    R = R_transpose.T

    return R, K, X0



R, K, X0 = ret_RKT(P)

print("K:\n", K, "\nR:\n", R, "\nX0:", X0, "\n")
```

We have successfully calculated the calibration matrix and the pose of the satellite by decomposing the projection matrix obtained by the DLT algorithm.  
Now let us verify how good our $P_{estimated}$ is.  

```python
#Get reprojected points
def get_points(P, X):
    x = P @ X.T
    x_normalized = x/x[2]
    x_euclidian = x_normalized.T[:, :2]
    return x_euclidian

#Get reprojection error
def get_err(x, X, P):
    x_projected = get_points(P, X)
    print("The initial image points are:\n", x[:, :2], "\nThe projected image points found through P and X are:\n", x_projected, "\n")
    error = np.linalg.norm(x[:, :2] - x_projected, axis = 1)
    average_distance_separation = np.mean(error) # np.mean(error ** 2)
    return average_distance_separation

print("Reprojection Error is:", get_err(x, X, P))
```

Clearly our overall algorithm is very accurate and can be employed in Indian geostationary satellites with other more features.
